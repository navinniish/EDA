```python
def clean_currency1(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('M', '').replace(',', ''))
    return(x)
```

```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('€', '').replace(',', ''))
    return(x)
```

```python
def clean_currency2(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('K', '').replace(',', ''))
    return(x)
```

```python
df1['Value'] = df1['Value'].apply(clean_currency)
df1['Value'] = df1['Value'].apply(clean_currency1)
df1['Value'] = df1['Value'].apply(clean_currency2)
df1['Value'] = df1['Value'].astype('float')
```

```python
df1['Wage'] = df1['Wage'].apply(clean_currency)
df1['Wage'] = df1['Wage'].apply(clean_currency1)
df1['Wage'] = df1['Wage'].apply(clean_currency2)
df1['Wage'] = df1['Wage'].astype('float')
```

```python
df1['Release Clause'] = df1['Release Clause'].apply(clean_currency)
df1['Release Clause'] = df1['Release Clause'].apply(clean_currency1)
df1['Release Clause'] = df1['Release Clause'].apply(clean_currency2)
df1['Release Clause'] = df1['Release Clause'].astype('float')
```

```python
def clean_height(x):
    """ If the value is a string, then remove 'inch' symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace("'", '.').replace(',', ''))
    return(x)
```

```python
df1['Height'] = df1['Height'].apply(clean_height)
df1['Height'] = df1['Height'].astype('float')
```

```python
def clean_weight(x):
    """ If the value is a string, then remove 'lbs' symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace("lbs", '').replace(',', ''))
    return(x)
```

```python
df1['Weight'] = df1['Weight'].apply(clean_weight)
df1['Weight'] = df1['Weight'].astype('float')
```

```python
df1.Joined = pd.to_datetime(df1.Joined)
df1["Joined"] = df1.Joined.dt.year
df1['Joined'] = df1.Joined.fillna(0).astype(int)
```

```python
df1['Contract Valid Until'] = pd.to_datetime(df1['Contract Valid Until'])
```
